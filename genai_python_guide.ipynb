## 3. Example: Prompting GenAI for Python Code

Suppose you are working with a dataset of global temperature anomalies in a CSV file and you want to quickly plot it in Python. Here’s a short example of how you might prompt a GenAI system.

**Prompt**:
```
I have a CSV file containing global monthly temperature anomalies with columns "Year", "Month", and "Temperature_Anomaly".
1. Write a Python function to load this file into a pandas DataFrame.
2. Show how to plot the temperature anomalies over time using matplotlib or seaborn.
3. Explain how the code works line by line.
```

**Potential GenAI Response** (annotated and verified by you afterward):

````python
import pandas as pd
import matplotlib.pyplot as plt

def load_temperature_data(csv_path):
    """
    Loads the temperature anomaly data into a pandas DataFrame.
    Args:
        csv_path (str): Path to the CSV file.
    Returns:
        pd.DataFrame: DataFrame containing the temperature anomalies.
    """
    df = pd.read_csv(csv_path)
    return df

def plot_temperature_anomalies(df):
    """
    Plots temperature anomalies over time.
    Args:
        df (pd.DataFrame): DataFrame with columns 'Year', 'Month', and 'Temperature_Anomaly'.
    """
    # Create a datetime column for plotting
    df['Date'] = pd.to_datetime(df[['Year', 'Month']].assign(DAY=1))
    plt.figure(figsize=(10, 5))
    plt.plot(df['Date'], df['Temperature_Anomaly'], label='Temperature Anomaly')
    plt.xlabel('Date')
    plt.ylabel('Temperature Anomaly (°C)')
    plt.title('Global Monthly Temperature Anomalies')
    plt.legend()
    plt.show()

# Example usage (uncomment if you have the data file):
# df = load_temperature_data("global_temp_anomalies.csv")
# plot_temperature_anomalies(df)
````

**Explanation**:

1. We import `pandas` and `matplotlib` to handle data and visualization.  
2. `load_temperature_data` reads the CSV using `pd.read_csv` and returns a DataFrame.  
3. `plot_temperature_anomalies` creates a `Date` column by combining Year and Month.  
4. We plot the anomalies over time and set labels and titles to make it clear.

## 4. Beyond the Basics: Prompt Engineering Strategies

1. **Use Step-by-Step Queries**:  
   Break down your requests rather than asking for everything at once. For example:  
   - “Write a Python function that reads a CSV file.”  
   - “Now show how to plot the data using matplotlib.”  
   - “Explain how to annotate the plot with a line for the mean anomaly.”  

2. **Ask for Explanations or Alternatives**:  
   - “Provide an alternative approach using xarray for NetCDF files.”  
   - “Explain each line of the code so I understand how it works.”  

3. **Iterate**:  
   - If the first result has errors or lacks clarity, ask the model to refine or correct the code.

## 5. Verifying & Documenting AI-Generated Code

### 5.1 Testing & Quality Control

- **Test locally**: Run the code in a local environment or a Jupyter notebook cell to ensure it works with your specific dataset.  
- **Check logic**: Make sure the code makes sense. AI can sometimes produce code that “looks right” but has logical errors.  
- **Use version control**: Tools like Git can help you compare changes and keep track of when and where AI-generated snippets are used.

### 5.2 Transparency & Academic Integrity

In academic or scientific contexts, be explicit about your sources. For instance, in the comments or documentation of your notebook, note:

> “Code in this section was generated with the assistance of [Model Name / ChatGPT / Bing Chat / etc.] on [Date].”

This transparency allows future readers (and your instructors) to understand how the code was produced, and it adheres to ethical standards of attribution.

## 6. Extended Example: Using GenAI with Earth-Science Libraries

GenAI can assist with more specialized libraries commonly used by Earth scientists. For example:

**Prompt**:
```
Write a Python script that opens a netCDF file using xarray, prints its metadata,
calculates the global average of a variable named 'surface_temp', and saves
the result to a new netCDF file. Provide explanations.
```

**Potential GenAI Response** (verify & adapt as needed):

````python
import xarray as xr

def process_surface_temperature(input_file, output_file):
    """
    Opens a netCDF file, calculates a global average of 'surface_temp', 
    and saves the result to a new netCDF file.

    Args:
        input_file (str): Path to the input netCDF file.
        output_file (str): Path to the output netCDF file.
    """
    # Open dataset
    ds = xr.open_dataset(input_file)
    
    # Print metadata
    print("Dataset information:")
    print(ds)
    
    # Calculate global average of surface_temp
    # Assumes 'lat' and 'lon' dimensions and that 'surface_temp' is the variable name
    # Also assumes ds.cell_area if you want area-weighted means.
    # For simplicity, let's do a naive mean here:
    mean_temp = ds.surface_temp.mean(dim=("lat", "lon"))
    
    # Convert to a dataset and save
    mean_temp_ds = mean_temp.to_dataset(name="global_mean_surface_temp")
    mean_temp_ds.to_netcdf(output_file)

# Example usage (uncomment if you have the data file):
# process_surface_temperature("in_file.nc", "out_file.nc")
````

**Explanation**:

1. **Library import**: `xarray` is a library specialized for labeled multi-dimensional arrays, often used for netCDF data in Earth science.  
2. **Opening dataset**: We use `xr.open_dataset` to open the netCDF file.  
3. **Metadata**: Printing the dataset reveals dimension sizes, coordinates, variable names, and attributes.  
4. **Mean temperature**: We take the naive mean across `lat` and `lon`. For area-weighted means, you’d use something like `ds.surface_temp.weighted(ds.cell_area).mean(dim=('lat', 'lon'))`.  
5. **Saving results**: We convert the `mean_temp` DataArray into a Dataset and save it to a new netCDF file.

> **Note**: This code snippet might require modifications (e.g., verifying that `cell_area` exists and has correct units). Always adapt and verify AI-generated code for your specific data structure.
